```python
def power(base, exponent):
    print(base ** exponent)
    return base ** exponent

power(2, 3)
```

```python
def grade_percentage(score: int, total: int) -> str:
    """Calculate the grade percentage."""
    
    percentage = (score/total) * 100
    if percentage >= 90:
        return "A"
    elif percentage >= 80:
        return "B"
    elif percentage >= 70:
        return "C"
    elif percentage >= 60:
        return "D"
    else:
        return "F"


```

```python
grade_percentage(80,100)
```
