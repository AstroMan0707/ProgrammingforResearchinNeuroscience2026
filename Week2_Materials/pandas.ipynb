```python
import numpy as np
import pandas as pd

series_1 = ['Delaware', 'Georgia', 'New Hampshire', 'Tennessee', 'Arkansas']
# np.array(series_1)

pd.Series(data = series_1)
```

```python
series_2 = [1, 4, 9, 16, 25] 

pd.Series(data = series_2)
```

```python
pd.Series(data = series_1, index = series_2)
```

```python
pd.Series(series_1, series_2)
```

```python
d = {'1': 'Delaware',
        '4': 'Georgia',
        '9': 'New Hampshire',
        '16': 'Tennessee', 
        '25': 'Arkansas'}
type(d)
```

```python
pd.Series(d) # you can pass in a dictionary
```

```python
my_series = pd.Series(d) # you can pass in a dictionary
my_series['16'] # index series with key
```
